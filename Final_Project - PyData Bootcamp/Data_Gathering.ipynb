### Importing all data libraries

```python
from bs4 import BeautifulSoup as bs
import requests as req
import pandas as pd
import numpy as np
```

```python
yplaza_url = 'https://www.yarnplaza.com/yarn/brands'
results = req.get(yplaza_url)
results
```

###### Thus our request has succeeded so we can now parse

```python
yarn = bs(results.content, 'html.parser')
yarn
```

```python
brands = yarn.find_all('h2', class_ = 'productlist-title')
brands = [b.text.strip() for b in brands]
brands
dmc= brands[1]
dmc
```

```python
dmc_url = 'https://www.yarnplaza.com/yarn/dmc'
r_dmc = req.get(dmc_url)
r_dmc
```

```python
y_dmc = bs(r_dmc.content, 'html.parser')
y_dmc
```

```python
dmc_nxl = y_dmc.find_all('div', class_ = 'productlist-topfx')
dmc_nxl = [nx.text.strip() for nx in dmc_nxl]
nxl = dmc_nxl[10][4:]
nxl
```

```python
nxl_url = 'https://www.yarnplaza.com/product/4216/dmc-natura-xl.html'
r_nxl = req.get(nxl_url)
r_nxl
```

```python
y_nxl = bs(r_nxl.content, 'html.parser')
y_nxl
```

##### for the price, composition and needle price

```python
nxl_p = y_nxl.find_all('span', class_= 'product-price-amount')
nxl_p1 = [p.text.strip() for p in nxl_p]
nxl_price = print(nxl_p1[0])
nxl_price
```

```python
## Converting to a float data type before rounding
nxl_price2 = float(8.53 if nxl_price is None else 0)
nxl_price2 = round(nxl_price2)
print('Price for Natura XL is €',nxl_price2 )
```

```python
nxl_c = y_nxl.find_all('td')
nxl_c1 = [c.text.strip() for c in nxl_c]
nxl_com = nxl_c1[13]
print('Composition for Natura XL is', nxl_com )
```

```python
nxl_n = y_nxl.find_all('td')
nxl_n1 = [n.text.strip() for n in nxl_n]
nxl_ns = nxl_n1[15]
print('The needle size for Natura XL is', nxl_ns)
```
